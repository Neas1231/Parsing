```python
from bs4 import BeautifulSoup  # Основная библиотека для работы с html кодом
import requests  # Встроенная библиотека python получения ответов от сайта
url = 'https://shikimori.one/animes'
page = requests.get(url)  # Получаем запрос от сайта с которого хотим парсить
print(page.status_code,end='\n\n')
print(page.text)
```

Определяем user-agent для нашего запроса, ведь в ином случае сайт не хочет пропускать наш request без этого параметра

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 YaBrowser/24.1.0.0 Safari/537.36'}
page = requests.get(url, headers=headers)
print(page.status_code)
```

```python
print(page.text)
```

```python
soup = BeautifulSoup(page.text, "html.parser")
print(soup.prettify())
```

```python
soup.find_all('a', class_='cover')[0]['href']
```

```python
urls_data = []
a_tags = soup.find_all('a', class_='cover')
for a in a_tags:
    print(a['href'])
    urls_data.append(a['href'])
```

```python
urls_data
```

```python
import pickle
with open('./urls_data.pickle','wb') as f:
    pickle.dump(urls_data,f)
```

```python
with open('./urls_data.pickle','rb') as f:
    urls_data = pickle.load(f)
    print(urls_data)
```
